```python
import pandas as pd
import numpy as np
```

```python
# Read transposed csv
arma3t = pd.read_csv('./arma3.csv', header=None).T
```

```python
#arma2_np = arma2t.to_numpy()
print(arma3t.shape)
```

```python
# List of timestamp to insert
# Change period according to length of the time series
datelist = pd.date_range(start='1/1/2022', freq='1H', tz='UTC', periods=100)
#print(type(datelist))

# Insert timestamp as the first column to transposed csv
arma3t.insert(0, "Timestamp", datelist)
```
